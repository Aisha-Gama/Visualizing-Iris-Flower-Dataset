```python
fig = plt.figure(figsize=(10,5))
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'])
plt.ylabel('SepalLengthCm')
plt.xlabel('SepalWidthCm')
plt.title('Scatter Plot')
plt.show()
```

```python
fig = plt.figure(figsize=(10,5))
plt.scatter(df['SepalLengthCm'], df['SepalWidthCm'], color = 'red')
plt.ylabel('PetalLengthCm')
plt.xlabel('PetalWidthCm')
plt.title('Scatter Plot')
plt.show()
```

```python
setosa_df = df[df['Species']== 'Iris-setosa']
setosa_df.head(5)
```

```python
color_map = {'Iris-setosa': 'red', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}

fig, ax = plt.subplots(figsize=(10, 5))

for species, group in df.groupby('Species'):
 plt.scatter(group['PetalLengthCm'], group['PetalWidthCm'],
               color=color_map[species], label=species)
ax.set_ylabel('PetalLengthCm')
ax.set_xlabel('PetalWidthCm')
ax.legend()

plt.show()
```

```python
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(figsize=(10, 6))

sns.histplot(data=df, x="PetalLengthCm", hue="Species", multiple="stack", ax=ax)
ax.set_xlabel("Petal Length (cm)")
ax.set_ylabel("Count")
ax.set_title("Distribution of Petal Length by Species")

plt.show()
```

```python
color_map = {'Iris-setosa': 'red', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}

fig, ax = plt.subplots(figsize=(10, 5))

for species, group in df.groupby('Species'):
 plt.scatter(group['SepalLengthCm'], group['SepalWidthCm'],
               color=color_map[species], label=species)
ax.set_ylabel('SepalLengthCm')
ax.set_xlabel('SepalWidthCm')
ax.legend()

plt.show()
```

Conclusion:

The data set has 150 datas, three species of Iris: setosa, versicolor, and virginica with 50 samples for each species.

Columns:Six; (Id, sepal length, sepal width, petal length, petal width and species).

Features: sepal length, sepal width, petal length, and petal width



From the analysis it was seen that:

-Iris-Setosa is easily separable from the others with smaller petals and sepals, while Iris-Versicolor and Iris-Virginica are more similar with overlaps, though Iris-virginica has the biggest petals Iris-versicolor has the largest sepals.

-The species are less separable with multiple ovelaps based on sepal length and width.

-Petal Length and Petal Width are the most informative features.
